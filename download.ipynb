```python
import os
import json
import re
import cv2
from datetime import datetime

import pandas as pd

from scraper import download_image
```

```python
df = pd.read_csv(r'C:\Users\feljo\Documents\satellite-imagery-downloader\src\Map_of_the_Worlds_Best_Surf_Spots.csv')
df
```

```python
config = {
        'url': 'https://mt.google.com/vt/lyrs=s&x={x}&y={y}&z={z}',
        'tile_size': 256,
        'tile_format': 'jpg',
        'dir': r'C:\Users\feljo\Documents\surf-finder\images',
        'headers': {
            'cache-control': 'max-age=0',
            'sec-ch-ua': '" Not A;Brand";v="99", "Chromium";v="99", "Google Chrome";v="99"',
            'sec-ch-ua-mobile': '?0',
            'sec-ch-ua-platform': '"Windows"',
            'sec-fetch-dest': 'document',
            'sec-fetch-mode': 'navigate',
            'sec-fetch-site': 'none',
            'sec-fetch-user': '?1',
            'upgrade-insecure-requests': '1',
            'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.82 Safari/537.36'
        },
    }
```

```python
for i in range(len(df[0:10])):

    location_name = df.loc[i]['Name']
    lon1, lat1 = df.loc[i]['X'] - 0.0018, df.loc[i]['Y'] + 0.0018
    lon2, lat2 = lon1 + 0.0018, lat1 - 0.0018

    img = download_image(
        lat1 = lat1,
        lon1 = lon1,
        lat2 = lat2,
        lon2 = lon2,
        zoom = 22,
        url = config['url'],
        headers = config['headers'],
        tile_size = config['tile_size'],
        channels = 3 
    )

    timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
    name = f'img_{timestamp}_{lat1}_{lon1}_{lat2}_{lon2}_{location_name}.png'
    cv2.imwrite(os.path.join(config['dir'], name), img)
    print(f'{location_name} saved as file {name}')
```
