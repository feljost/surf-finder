```python
import pandas as pd

df = pd.read_csv(r'C:\Users\feljo\Documents\surf-finder\Map_of_the_Worlds_Best_Surf_Spots.csv')

df
```

```python

for i in tqdm(range(len(df))):

    location_name = df.loc[i]['Name']

    top_left, bottom_right = calculate_box(df.loc[i]['Y'], df.loc[i]['X'], 0.25)


    img = download_image(
        lat1 = top_left[0],
        lon1 = top_left[1],
        lat2 = bottom_right[0],
        lon2 = bottom_right[1],
        zoom = 22,
        url = config['url'],
        headers = config['headers'],
        tile_size = 256, #config['tile_size'],
        channels = 3,
        resolution_scaling = 0.12 # original pixels are way less than the quality of the image
    )

    timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
    name = f'img_{timestamp}_{location_name}.png'
    cv2.imwrite(os.path.join(config['dir'], name), img)
    
```
